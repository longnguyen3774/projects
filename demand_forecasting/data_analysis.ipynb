Phân bố tuổi (age) theo giới tính (gender) của khách hàng

```python
df_customers['date_of_birth'] = pd.to_datetime(df_customers['date_of_birth'])
df_customers['age'] = 2026 - df_customers['date_of_birth'].dt.year
```

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))

plot = sns.boxplot(
    data=df_customers,
    x='gender',
    y='age',
    palette='Set2',
    hue='gender',
    legend=False
)

plt.title('Phân bố của độ tuổi khách hàng theo giới tính', fontsize=16, fontweight='bold')
plt.xlabel('Giới tính', fontsize=12)
plt.ylabel('Tuổi', fontsize=12)

plt.tight_layout()
plt.show()
```

Nhận xét:
- Khách hàng chủ yếu là người trẻ, có độ tuổi từ 20 đến 40
- Không có sự khác nhau tuổi tác giữa các nhóm giới tính

Tổng số sản phẩm theo danh mục

```python
df_plot = df_products['category'].value_counts().reset_index()
df_plot.columns = ['category', 'total_products']

df_plot = df_plot.sort_values(by='total_products', ascending=False)

plt.figure(figsize=(12, 7))
sns.set_theme(style="whitegrid")

plot = sns.barplot(
    data=df_plot,
    x='total_products',
    y='category',
    palette='magma',
    hue='category',
    legend=False
)

plt.title('Tổng số sản phẩm theo danh mục', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Số lượng sản phẩm', fontsize=12)
plt.ylabel('Danh mục', fontsize=12)

for i, v in enumerate(df_plot['total_products']):
    plt.text(v + 0.5, i, f'{int(v):,}', va='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
df_plot = df_products['sub_category'].value_counts().reset_index()
df_plot.columns = ['sub_category', 'total_products']

df_plot = df_plot.sort_values(by='total_products', ascending=False)

plt.figure(figsize=(12, 7))
sns.set_theme(style="whitegrid")

plot = sns.barplot(
    data=df_plot,
    x='total_products',
    y='sub_category',
    palette='magma',
    hue='sub_category',
    legend=False
)

plt.title('Tổng số sản phẩm theo danh mục phụ', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Số lượng sản phẩm', fontsize=12)
plt.ylabel('Danh mục phụ', fontsize=12)

for i, v in enumerate(df_plot['total_products']):
    plt.text(v + 0.5, i, f'{int(v):,}', va='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

Tỷ lệ các phương thức thanh toán

```python
data_pie = df_transactions[df_transactions['transaction_type'] == 'Sale']['payment_method'].value_counts()
labels = data_pie.index
values = data_pie.values

colors = sns.color_palette('viridis', len(labels))

fig, ax = plt.subplots(figsize=(10, 8), subplot_kw=dict(aspect="equal"))

wedges, texts, autotexts = ax.pie(
    values,
    labels=labels,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    pctdistance=0.85,
    explode=[0.03] * len(labels),
    textprops={'fontsize': 12, 'fontweight': 'bold'}
)

centre_circle = plt.Circle((0,0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.setp(autotexts, size=10, color="white")

ax.legend(wedges, labels, title="Phương thức", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

plt.title('Cấu trúc phương thức thanh toán', fontsize=18, fontweight='bold', pad=20)

plt.tight_layout()
plt.show()
```

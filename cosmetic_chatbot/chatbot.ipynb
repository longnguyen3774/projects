```python
from transformers import AutoTokenizer, AutoModel, AutoModelForCausalLM
import torch
from pyvi import ViTokenizer

phobert_tokenizer = AutoTokenizer.from_pretrained('vinai/phobert-base')

phobert_model = AutoModel.from_pretrained('vinai/phobert-base')

phobert_model.eval()

llm_model_id = 'mistralai/Mistral-7B-Instruct-v0.2'

llm_tokenizer = AutoTokenizer.from_pretrained(llm_model_id)

llm_model = AutoModelForCausalLM.from_pretrained(
    llm_model_id,
    device_map='auto',
    torch_dtype=torch.float16
)
```

```python
import json
import uuid
import torch
from tqdm import tqdm
from qdrant_client import QdrantClient
from qdrant_client.http.models import PointStruct, VectorParams, Distance
from transformers import AutoModel, AutoTokenizer
from underthesea import word_tokenize

if COLLECTION_NAME not in [col.name for col in client.get_collections().collections]:
    client.create_collection(
        collection_name=COLLECTION_NAME,
        vectors_config=VectorParams(size=768, distance=Distance.COSINE)
    )
    print("✅ Đã tạo collection.")

def get_phobert_embedding(text):
    text = word_tokenize(text, format="text")
    inputs = phobert_tokenizer(text, return_tensors="pt", truncation=True, max_length=256)
    with torch.no_grad():
        outputs = phobert_model(**inputs)
        embeddings = outputs.last_hidden_state.mean(dim=1)  # Lấy trung bình các token
        return embeddings.squeeze().tolist()

with open("my_pham.json", "r", encoding="utf-8") as f:
    data = json.load(f)

points = []
for item in tqdm(data):
    try:
        content = f"{item.get('Tên sản phẩm', '')}. Giá: {item.get('Giá', '')}. {item.get('Thông tin sản phẩm', '')}"
        embedding = get_phobert_embedding(content)
        point = PointStruct(
            id=str(uuid.uuid4()),
            vector=embedding,
            payload=item
        )
        points.append(point)
    except Exception as e:
        print(f"[Lỗi] {item.get('Tên sản phẩm', '')} - {e}")

client.upsert(collection_name=COLLECTION_NAME, points=points)
print("✅ Đã lưu toàn bộ sản phẩm vào Qdrant.")
```

```python
def get_question_embedding(text):
    text = ViTokenizer.tokenize(text)

    inputs = phobert_tokenizer(
        text,
        return_tensors='pt',
        truncation=True,
        max_length=256 
    )

    with torch.no_grad():
        outputs = phobert_model(**inputs)
        
        last_hidden = outputs.last_hidden_state

        attention_mask = inputs['attention_mask']

        mask = attention_mask.unsqueeze(-1).expand(last_hidden.size()).float()

        masked = last_hidden * mask

        summed = masked.sum(dim=1)

        count = mask.sum(dim=1)

        mean_pooled = summed / count

        return mean_pooled.squeeze().tolist()
```

```python
def generate_response(prompt):
    inputs = llm_tokenizer(prompt, return_tensors='pt').to(llm_model.device)

    input_ids_len = inputs['input_ids'].shape[1]

    with torch.no_grad():
        outputs = llm_model.generate(
            **inputs,
            max_new_tokens=1024,
            temperature=0.7,
            top_p=0.9,
            do_sample=True,
            pad_token_id=llm_tokenizer.eos_token_id
        )

    generated_ids = outputs[0][input_ids_len:]

    return llm_tokenizer.decode(generated_ids, skip_special_tokens=True).strip()
```

```python
def build_prompt(question, results):
    context = ""
    for i, item in enumerate(results, start=1):
        name = item.get("Tên sản phẩm", "Không rõ")
        price = item.get("Giá", "Không có")
        info = item.get("Thông tin sản phẩm", "Không có mô tả")
        context += f"{i}. {name}\n   Giá: {price}\n   Mô tả: {info[:300]}...\n\n"

    return f"""Khách hàng hỏi: "{question}"

Dưới đây là danh sách sản phẩm liên quan:

{context}

Hãy tư vấn cho khách hàng một cách thân thiện, tự nhiên như đang trò chuyện.
Chỉ ra 2-3 sản phẩm phù hợp nhất, giải thích lý do chọn lựa.
Nếu thấy cần thiết, có thể đưa ra lời khuyên chung.

Trả lời bằng tiếng Việt, văn phong gần gũi và dễ hiểu."""
```

```python
def search_products(query_vector, top_k=TOP_K):
    results = client.search(
        collection_name=COLLECTION_NAME,
        query_vector=query_vector,
        limit=top_k
    )
    return [r.payload for r in results]
```

```python
def chatbot(question):
    vector = get_question_embedding(question)

    results = search_products(vector, top_k=3)
    if not results:
        return "❌ Không tìm thấy sản phẩm phù hợp."

    prompt = build_prompt(question, results)

    return generate_response(prompt)
```

```python
import os
from huggingface_hub import InferenceClient
from google.colab import userdata

HF_TOKEN = userdata.get('HF_TOKEN')

hf_client = InferenceClient(
    provider="featherless-ai",
    api_key=HF_TOKEN,
)
```

```python
def chatbot_online(question):
    vector = get_question_embedding(question)

    results = search_products(vector, top_k=3)
    if not results:
        return "❌ Không tìm thấy sản phẩm phù hợp."

    prompt = build_prompt(question, results)

    completion = hf_client.chat.completions.create(
        model="mistralai/Mistral-7B-Instruct-v0.2",
        messages=[
            {"role": "system", "content": "Bạn là một trợ lý gợi ý sản phẩm, thân thiện, nói tiếng Việt."},
            {"role": "user", "content": prompt}
        ],
        temperature=0.7,
        max_tokens=1024,
        top_p=0.9
    )

    return completion.choices[0].message.content.strip()
```

```python
question = input("❓ Nhập câu hỏi của bạn: ")
response = chatbot_online(question)
print("\n🤖 Trợ lý gợi ý:\n")
print(response)
```

```python
from fastapi import FastAPI, Form, Request
from fastapi.responses import HTMLResponse
from fastapi.templating import Jinja2Templates
from starlette.middleware.sessions import SessionMiddleware

app = FastAPI()
app.add_middleware(SessionMiddleware, secret_key="secret")

templates = Jinja2Templates(directory="templates")

@app.get("/", response_class=HTMLResponse)
async def form_get(request: Request):
    history = request.session.get("history", [])
    return templates.TemplateResponse("chat.html", {
        "request": request,
        "history": history
    })

@app.post("/", response_class=HTMLResponse)
async def form_post(request: Request, question: str = Form(...)):
    reply = chatbot_online(question)

    history = request.session.get("history", [])
    history.append({"sender": "user", "message": question})
    history.append({"sender": "bot", "message": reply})
    request.session["history"] = history

    return templates.TemplateResponse("chat.html", {
        "request": request,
        "history": history
    })
```
